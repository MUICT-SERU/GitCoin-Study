```
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

filtered_data = allBounty.loc[allBounty['token_name'].isin(['ETH']), ['token_value_time_peg', 'token_value_in_usdt', 'is_success']]
filtered_data['token_value_time_peg'] = pd.to_datetime(filtered_data['token_value_time_peg'])  # Convert to datetime format
filtered_data = filtered_data.set_index('token_value_time_peg')  # Set the index to the time column


filtered_data2 = allBounty.loc[allBounty['token_name'].isin(['ETH']), ['token_value_time_peg', 'token_value_in_usdt']]
filtered_data2['token_value_time_peg'] = pd.to_datetime(filtered_data2['token_value_time_peg'])  # Convert to datetime format
filtered_data2 = filtered_data2.sort_values('token_value_time_peg') 

plt.figure(figsize=(10, 6))

# Plot token_value_in_usdt
plt.plot(filtered_data2['token_value_time_peg'], filtered_data2['token_value_in_usdt'])
plt.xlabel('Month and Year')
plt.ylabel('Token Value in USDT')

# Calculate count of 'True' values grouped by month
success_count_monthly = filtered_data.resample('M')['is_success'].apply(lambda x: (x == True).sum())

# Create secondary y-axis for success count
ax2 = plt.twinx()

# Plot success count as a bar chart
ax2.bar(success_count_monthly.index, success_count_monthly.values, alpha=0.5, color='green', label='Success Count')
ax2.set_ylabel('No. of success issues')

# plt.title('Token Value in USDT and Success Count over Time (Grouped by Month)')
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=12)


# Display legends for both plots
lines, labels = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
# plt.legend(lines + lines2, labels + labels2, loc='upper left')

plt.grid(True)
plt.show()
```
